```python
import catboost
from catboost import CatBoostClassifier, Pool
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from pandas import Series, DataFrame
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("result.csv")
```

```python
df=df.drop('Unnamed: 0',axis=1)
df=df.drop(['salary'], axis=1)
```

Выделяю наиболее частые категории, ввиду большого количества категорий в целевой метке

```python
df=df.loc[(df['job-name'] == 'Швея') | (df['job-name'] == 'Специалист') | (df['job-name'] == 'Уборщик производственных и служебных помещений') 
       | (df['job-name'] == 'Врач ') | (df['job-name'] == 'Водитель автомобиля') | (df['job-name'] == 'Подсобный рабочий')
      | (df['job-name'] == 'Медицинская сестра') | (df['job-name'] == 'Повар') | (df['job-name'] == 'Технолог')
       | (df['job-name'] == 'Генеральный директор предприятия') | (df['job-name'] == 'Токарь') | (df['job-name'] == 'Закройщик')
      | (df['job-name'] == 'Дворник')]
```

Трансформация категориальных признаков в числа

```python
le = LabelEncoder()
encode_feature = le.fit_transform(df["job-name"])
```

```python
le = LabelEncoder()
x1= le.fit_transform(df["employment"])
```

```python
le = LabelEncoder()
x2= le.fit_transform(df["schedule"])
```

Присваиваю набору данных новые признаки

```python
df["x1"]=x1
df["x2"]=x2
```

```python
df=df.drop(["employment"], axis=1)
df=df.drop(["schedule"], axis=1)
```

```python
df=df.drop(['job-name'], axis=1)
```

```python
Делю выборку на тест и трейн в соотношении 0.3 и 0.7
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, encode_feature, 
                 test_size=0.1, stratify= encode_feature)
```

```python
train_pool= Pool(data=X_train, label= y_train)

test_pool= Pool(data=X_test, label= y_test)             

model=CatBoostClassifier(iterations=1000, depth=16, learning_rate=0.1, 
                         loss_function='MultiClass', eval_metric= 'Accuracy', logging_level='verbose')

model.fit(train_pool, eval_set=test_pool, logging_level='Verbose')

pred= model.predict(data=test_pool, prediction_type='Class')

acc_test=accuracy_score(y_test, pred)

pred= model.predict(data=train_pool, prediction_type='Class')

acc_train=accuracy_score(y_train, pred)
```
